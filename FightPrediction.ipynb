```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
fighter_df = pd.read_csv('raw_fighter_details.csv')
fight_df = pd.read_csv('raw_total_fight_data.csv',sep=';')
```

```python
fighter_df.head()
```

```python
fight_df.head()
```

```python
fight_df = fight_df[['R_fighter','B_fighter','Winner']]
```

```python
df = fight_df.merge(fighter_df, how='left', left_on='R_fighter', right_on='fighter_name')
df = df.merge(fighter_df, how='left', left_on='B_fighter', right_on='fighter_name',suffixes=('_Red', '_Blue'))
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Winner','DOB_Red','DOB_Blue','Height_Red','Height_Blue','Weight_Blue','Weight_Red'])
```

```python
df.isna().sum()
```

```python
cols = ['Reach_Red','Reach_Blue','Stance_Red','Stance_Blue']
for i in cols:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df['Height_Red'] = df['Height_Red'].str.replace("'",".")
df['Height_Red'] = df['Height_Red'].str.replace('"','')
df['Height_Red'] = df['Height_Red'].str.replace(" ","")
df['Height_Red'] = df['Height_Red'].astype(float)
```

```python
df['Weight_Red'] = df['Weight_Red'].str.replace(' lbs.','').astype(int)
```

```python
df['Reach_Red'] = df['Reach_Red'].str.replace('"','').astype(int)
df['Height_Blue'] = df['Height_Blue'].str.replace("'",".")
df['Height_Blue'] = df['Height_Blue'].str.replace('"','')
df['Height_Blue'] = df['Height_Blue'].str.replace(" ","")
df['Height_Blue'] = df['Height_Blue'].astype(float)
df['Reach_Blue'] = df['Reach_Blue'].str.replace('"','').astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Str_Acc_Red'] = df['Str_Acc_Red'].str.replace('%','').astype(int)
df['TD_Acc_Red'] = df['TD_Acc_Red'].str.replace('%','').astype(int)
df['TD_Def_Red'] = df['TD_Def_Red'].str.replace('%','').astype(int)
df['Weight_Blue'] = df['Weight_Blue'].str.replace(' lbs.','').astype(int)
df['Str_Acc_Blue'] = df['Str_Acc_Blue'].str.replace('%','').astype(int)
df['TD_Acc_Blue'] = df['TD_Acc_Blue'].str.replace('%','').astype(int)
df['TD_Def_Blue'] = df['TD_Def_Blue'].str.replace('%','').astype(int)
```

```python
df.info()
```

```python
df['Str_Def_Red'] =  df['Str_Def_Red'].str.replace('%','').astype(int)
```

```python
df['Str_Def_Blue'] =  df['Str_Def_Blue'].str.replace('%','').astype(int)
```

```python
df.info()
```

```python
df = df.drop(columns=['R_fighter','B_fighter','DOB_Red','DOB_Blue'])
```

```python
df.head()
```

```python
def winner(x,red,blue):
    if red in x:
        return 1
    elif blue in x:
        return 0
```

```python
df['winner_color'] = df[['Winner','fighter_name_Red','fighter_name_Blue']].apply(lambda x: winner(x['Winner'], x['fighter_name_Red'], x['fighter_name_Blue']), axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Stance_Red'] = le.fit_transform(df['Stance_Red'])
df['Stance_Blue'] = le.fit_transform(df['Stance_Blue'])
```

```python
df_red = df.iloc[:,1:14]
df_blue = df.iloc[:,14:-1]

df_red.to_csv('df_red.csv')
df_blue.to_csv('df_blue.csv')
```

```python
df = df.drop(['fighter_name_Blue','fighter_name_Red','Winner'],axis=1)
```

```python
df.head()
```

```python
df.head()
```

```python
X = df.drop(['winner_color'],axis=1)
y = df['winner_color']

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42, max_depth=100)
rf = RandomForestClassifier(random_state=42)
lr = LogisticRegression(random_state=42, max_iter=15000)

#rf.fit(X_train,y_train)
lr.fit(X_train,y_train)
#dt.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```
